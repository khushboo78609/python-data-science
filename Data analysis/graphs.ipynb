```python
import numpy as np
import matplotlib.pyplot as plt
```

# Chart and graphs
- Bar chart: used to compare the values of different categories of data
- Line chart: used to show trends over time
- Area chart: used to show trends over time,but can also be used to show the different between two 
- Pie chart: used to show the proportion of each category of data
- Histogram: used to show the distribution of a numerical value
- Scatter plot: used to show the relationship between two numerical variables


```python
rainfall = np.array([ 399, 555, 464, 587, 559])
years = np.array([2019,2020,2021,2022,2023])
```

```python
# line plot
plt.plot(rainfall)
```

```python
plt.plot(years,rainfall)
```

```python
# configuring graph
plt.plot(years, rainfall)
plt.title('Rainfall per year', fontdict={'fontsize':15,'color':'green'})
plt.xlabel('Years')
plt.ylabel("Rainfall (cm)")
plt.xticks(years)
plt.hlines(y=500, xmin=2019 , xmax = 2023, color='red', linestyles='--')
plt.show()

```

```python

plt.figure(figsize=(10,4), facecolor='lightblue'),
x  = np.array(['Rohan','Rahul','Raj','Ramesh',  'Rakesh'])
y = np.array([60,80,90,60,50])
bars =plt.bar(x,y,
         color=['orange','pink','purple','blue'],
         alpha= .5,edgecolor='black',linewidth=1, hatch='/')
       
bars[2].set_hatch('o')
bars[2].set_color('red')

plt.title("Data")
plt.ylim(0,120)
for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2, 
             bar.get_height()+10,
             f"${bar.get_height()}k",
            fontsize=0, color='red',
            ha='center')
plt.show()
```
