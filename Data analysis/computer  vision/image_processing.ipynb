```python
from PIL import Image ,ImageDraw , ImageFont
import requests
```

```python
def download_image(url):
    response = requests.get(url, stream=True)
    return Image.open(response.raw)
```

```python
im1 = download_image('https://assets.pokemon.com/assets/cms2/img/pokedex/full/151.png')
im1
```

```python
# generated via code
im3 = Image.new('RGB',(300,300), color='red')
im4 = Image.new('RGB',(300,300), color='green')
im3.show()
im4.show()
```

```python
# writing an  image
im5 = Image.new('RGB', (300,300), color='red')
font = ImageFont.truetype('font.ttf',40)
font_sm = ImageFont.truetype('font.ttf',20)
draw = ImageDraw.Draw(im5)
draw.text(
    (150,150),
    text='Digipodium',
    fill=(255,155,0), 
    font = font,
    anchor='mm', # middle middle 
    align='center',
)
draw.text((150,100), "Team",(255,255,255), font,'ms')
draw.text((150,260), "generated via code",(0,0,0), font_sm,'mm')

im5.save('coded_image.png')
im5
```
